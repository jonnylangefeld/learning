A Dataframe consists always of an **index** (first column, just an integer count in this case) and **pandas series** which are the other columns. To select only one columns (one series or one vector as in R) just use [ ]:    

---
**respective R code:**

    head(iris['species'])

### Slicing
That's great! Now let's slice our data a bit. The Easiest way is to select multiple columns by Name is with a list in the [ ]. That makes two brackets (always remember: [ ] gives back a series and [ [ ] ] gives back a dataframe):    

---
**respective R code:**

    iris[c('sepal_length','sepal_length')]

```python
iris[['sepal_length','sepal_length']].head()
```

Also a similar function as in R for selecting rows and columns exists. Just use [`.ix[rows,cols].`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.ix.html) You can use ranges. A single ':' stands for all:    

---
**respective R code:**

    iris[,0:2]

```python
iris.ix[:,0:2].head()
```

[`.loc[]`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html) is also there for labled selection of columns, but you can also input boolean vectors such as `iris['species']==setosa` in this case    

---
**respective R code:**

    iris[iris['species']=='setosa',]
    
Notice the comma in R!

```python
iris.loc[iris['species']=='setosa'].head()
```

The same thing is possible with the [`.query()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.query.html) function:    

---
**respective R code:** (not quite but similar)

    iris[which(iris$species=='setosa'),]

```python
iris.query('species == "setosa"').head()
```

All in all there are hundreds of options for slicing dataframes. It's best to refer to the [offical documentation](http://pandas.pydata.org/pandas-docs/stable/indexing.html#different-choices-for-indexing). But a gut resource is also always the help() function:    

---
**respective R code:**

    ?which

```python
help(pd.DataFrame.query)
```

### Aggregation:
Pandas provides the [`.groupby()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html) function:

```python
g = iris.groupby(['species'])
g
```

As you see it's result is just another groupby object. So you want to apply any aggregate function.    

---
**respective R code:** (notice that R doesn't automatically recognizes Factors. Exclude them first.)

    aggregate(,iris[1:4], by=list(iris$species), FUN=sum, na.rm=TRUE)
    summary(iris)

```python
g.sum()
```

```python
g.describe()
```

### Append and Merge:
Let's [`.merge()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) two iris datasets:    

---
**respective R code:** 

    merge(iris, iris, by.x = 0, by.y = 0, suffixes = c('_left','_right'))

```python
pd.merge(iris, iris, how='inner', left_index=True, right_index=True, suffixes=('_left','_right')).head()
```

This was a merge on indices, but you can also refer to specific columns. It's explained in the documentation linked above.

Now for the sake of [`append()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.append.html) let's split our dataframe first as learned above:
